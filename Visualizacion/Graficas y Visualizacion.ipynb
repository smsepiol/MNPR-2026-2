## Box-Plot

Para mostrar la distribución de datos y detectar outliers, comparar distribuciones entre diferentes grupos.

```r
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Crear gráfico de caja
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Gráfico de Caja')
plt.xlabel('Grupos')
plt.ylabel('Valores')
plt.xticks([1, 2, 3], ['Grupo 1', 'Grupo 2', 'Grupo 3'])
plt.show()
```

## Grafica incorrecta

**No identificar la grafica que me ayuda a visualizar correctamente la informacion que tengo.**

```r
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 2]

plt.plot(categories, values)
plt.title('Gráfico de Líneas para Datos Categóricos')
plt.show()
```

```r
plt.bar(categories, values)
plt.title('Gráfico de Barras para Datos Categóricos')
plt.show()
```

## Falta de escala y etiquetas en los ejes

**No indicar la escala o etiquetas en los ejes pueden volver inutil mi grafico para una persona fuera de contexto**

```r
import numpy as np
import matplotlib.pyplot as plt

#  matriz 10 x 10 de enteros entre 1 y 50
data = np.random.randint(low=1, high=50, size=(5, 5))

# mapa de calor
plt.imshow(data, cmap='viridis')

# Eliminar etiquetas y ejes
plt.xticks([])
plt.yticks([])

plt.show()
```

## Escala truncada

**No mostrar la escala completa, puede alterar las "proporciones" de los datos, y dar pie a malinterpretaciones**

```r
import matplotlib.pyplot as plt
import numpy as np

x = ["A","B"]
y = np.array([90, 100])

plt.bar(x, y)
plt.ylim(85, 100)  # Escala truncada
plt.title('Gráfico con Escala Truncada')
plt.show()
```

```r
plt.bar(x, y)
plt.ylim(0, 100)  # Escala correcta
plt.title('Gráfico con Escala correcta')
plt.show()
```

## Graficos saturados

**Escoger la grafica correcta pero saturarla de "detalles" que no permitan visualizar la informacion**


```r
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

#grafica diferentes lineas
for i in range(20):
    y = np.sin(x + i) + np.random.normal(0, 0.1, x.shape)  # Seno con ruido normal
    plt.plot(x, y, label=f'Data {i}', alpha=0.5 )  # Usar alpha para hacer las líneas más transparentes

plt.title('Gráfico Saturado')
plt.xlabel('Este es un titulo exagerado para la variable X 1234058272849')
plt.ylabel('Este es un titulo exagerado para la variable Y 123405827')
plt.legend()
plt.grid()
plt.show()
```

## Graficos innecesariamente complejos

**Crear un grafico innecesariamente complejo para mis datos**

Por ejemplo, supongamos que tengo un conjunto de puntos en $\mathbb{R}^{2}$, y yo utilizo un grafico en $\mathbb{R}^{3}$, realmente era necesario utilizar un grafico mas complejo?



```r
import plotly.graph_objects as go
import numpy as np


x = np.random.rand(100)
y = np.random.rand(100)
z = np.zeros(100)  # z = 0 para todos los puntos

# Crear el gráfico 3D
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers')])

# Configurar el diseño del gráfico
fig.update_layout(
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ),
    title='Gráfico de puntos con z=0'
)

fig.show()
```

# **Ejemplo de visualizacion en Regresion**:

A continuacion un ejemplo de algunas graficas que pueden ser de nuestro interes dentro de un analisis de regresion.

Supongamos que se tiene un conjunto de datos, compuesto de dos variables numericas y una nominal, y se desea ajustar un modelo de regresion, como parte del analisis descriptivo de nuestros datos, nos interesa saber si existe una relacion lineal entre las variables, asi como la deteccion de **clusters** o outliers.

```r
import pandas as pd
import numpy as np


# Crear datos
n = 100
x1 = np.random.normal(10, 2, n)
group = np.random.choice(['A', 'B'], size=n)  # Variable nominal con 2 niveles

#definimos x2 convenientemente para que se generen 2 rectas de regresion diferentes
x2 = np.zeros(n)
for i in range(n):
    if group[i] == 'A':
        x2[i] = 2 * x1[i] + np.random.normal(0, 2)  # Recta de regresión para el grupo A
    else:
        x2[i] = 0.5 * x1[i] + np.random.normal(0, 5)  # Recta de regresión para el grupo B

data = pd.DataFrame({'x1': x1, 'x2': x2, 'group': group})
data['group'] = pd.Categorical(data['group'])

data.head()
data.info()
```

Ahora sabemos que un grafico de dispersion puede ayudarme a visualizar mis datos de forma correcta.

```r
# Visualizar los datos
plt.figure(figsize=(10, 6))
plt.scatter(data[data['group'] == 'A']['x1'], data[data['group'] == 'A']['x2'],
            color='blue', alpha=1, label='Grupo A', marker='o')

plt.scatter(data[data['group'] == 'B']['x1'], data[data['group'] == 'B']['x2'],
            color='orange', alpha=1, label='Grupo B', marker='v')


plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Gráfico de Dispersión')
plt.legend()
plt.show()
```

Intuitivamente podemos notar que la variable nominal separa a mi poblacion, por lo que se ajustan dos modelos diferentes. **Mas adelante formalizaremos esta idea.**

```r
import statsmodels.api as sm

# Ajustar la regresión para el grupo A
data_A = data[data['group'] == 'A']
X_A = sm.add_constant(data_A['x1'])  # Agregar constante (intercepto)
model_A = sm.OLS(data_A['x2'], X_A).fit()

# Ajustar la regresión para el grupo B
data_B = data[data['group'] == 'B']
X_B = sm.add_constant(data_B['x1'])  # Agregar constante (intercepto)
model_B = sm.OLS(data_B['x2'], X_B).fit()
```

```r

# malla de valores para recta de regresion
x1_range = np.linspace(data['x1'].min(), data['x1'].max(), 100)

# Obtener puntos de la recta de regresion y calcular intervalos de confianza para el grupo A
pred_A = model_A.get_prediction(sm.add_constant(x1_range))
pred_A_summary = pred_A.summary_frame(alpha=0.05)


# Obtener puntos de la recta de regresion y calcular intervalos de confianza para el grupo B
pred_B = model_B.get_prediction(sm.add_constant(x1_range))
pred_B_summary = pred_B.summary_frame(alpha=0.05)


plt.figure(figsize=(12, 6))

# Grupo A
#Grafica los puntos de los datos
plt.scatter(data_A['x1'], data_A['x2'], color='blue', label='Grupo A', alpha=0.5,marker='o')
#Grafica la recta de regresion A
plt.plot(x1_range, model_A.predict(sm.add_constant(x1_range)), color='blue', label='Regresión A', linewidth=2)
#Rellena el area entre los intervalos de confianza
plt.fill_between(x1_range, pred_A_summary['obs_ci_lower'], pred_A_summary['obs_ci_upper'], color='blue', alpha=0.2)

# Grupo B
plt.scatter(data_B['x1'], data_B['x2'], color='orange', label='Grupo B', alpha=0.5,marker='v')
plt.plot(x1_range, model_B.predict(sm.add_constant(x1_range)), color='orange', label='Regresión B', linewidth=2)
plt.fill_between(x1_range, pred_B_summary['obs_ci_lower'], pred_B_summary['obs_ci_upper'], color='orange', alpha=0.2)

# Configuración de la gráfica
plt.title('Regresión Lineal por Grupo')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.grid()
plt.show()
```
